2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
import seaborn as sns
waterfront = df['waterfront'].value_counts()
sns.barplot(x= waterfront.index,  y = waterfront.values)
plt.title('Распределение домов от наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Количество домов');
```

2.2 Изучите распределение этажей домов

```python
floor = df['floors'].value_counts()

name = floor.index
colom = floor.values

plt.pie(colom,autopct='%.1f%%', labels=name);
```

2.2 Изучите распределение состояния домов

```python
condition = df['condition'].value_counts()
sns.barplot(x= condition.index,  y = condition.values)
plt.title('Распределение состояния домов')
plt.xlabel('Оценка состояния')
plt.ylabel('Количество домов');
```

Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
import numpy as np
corr_m = df.corr()
corr_m = np.round(corr_m,1)
corr_m[np.abs(corr_m) < 0.3] = 0
corr_m
```

```python
sns.heatmap(corr_m, annot=True, cmap='coolwarm');
```
